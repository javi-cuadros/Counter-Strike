# Número de rondas de cada tipo por mapa

Bibliotecas necesarias para la ejecución

```python
import pyodbc 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Definición de parámetros para la conexión con la base de datos

```python
conn_str_esea = (
    r'Driver={SQL Server};'
    r'Server=localhost\MSSQLSERVER01;'
    r'Database=csgo_esea;'
    r'Trusted_Connection=yes;'
    )
cnxn = pyodbc.connect(conn_str_esea)
```

Definición de query de obtención de datos

```python
query_cache = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_cache'  group by map,round_type order by map,round_type"
query_dust = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_dust2'  group by map,round_type order by map,round_type"
query_overpass = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_overpass'  group by map,round_type order by map,round_type"
query_nuke = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_nuke'  group by map,round_type order by map,round_type"
query_train = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_train'  group by map,round_type order by map,round_type"
query_mirage = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_mirage'  group by map,round_type order by map,round_type"
query_inferno = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_inferno'  group by map,round_type order by map,round_type"
query_cobblestone = "SELECT round_type,count([round_type]) contador FROM [csgo_esea].[dbo].[meta] where map = 'de_cbble'  group by map,round_type order by map,round_type"
```

Carga de datos en dataframes

```python
data_cache = pd.read_sql_query(query_cache, cnxn)
data_dust = pd.read_sql_query(query_dust, cnxn)
data_overpass = pd.read_sql_query(query_overpass, cnxn)
data_nuke = pd.read_sql_query(query_nuke, cnxn)
data_train = pd.read_sql_query(query_train, cnxn)
data_mirage = pd.read_sql_query(query_mirage, cnxn)
data_inferno = pd.read_sql_query(query_inferno, cnxn)
data_cobblestone = pd.read_sql_query(query_cobblestone, cnxn)
```

```python
fig, ax = plt.subplots()
contador = data_cache['contador']
ax.set(title='Cache')
plt.pie(contador, labels=data_cache['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_dust['contador']
ax.set(title='Dust')
plt.pie(contador, labels=data_dust['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_overpass['contador']
ax.set(title='Overpass')
plt.pie(contador, labels=data_overpass['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_nuke['contador']
ax.set(title='Nuke')
plt.pie(contador, labels=data_nuke['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_train['contador']
ax.set(title='Train')
plt.pie(contador, labels=data_train['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_mirage['contador']
ax.set(title='Mirage')
plt.pie(contador, labels=data_mirage['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_inferno['contador']
ax.set(title='Inferno')
plt.pie(contador, labels=data_inferno['round_type'], autopct="%0.1f %%")
plt.show()

fig, ax = plt.subplots()
contador = data_cobblestone['contador']
ax.set(title='Cobblestone')
plt.pie(contador, labels=data_cobblestone['round_type'], autopct="%0.1f %%")
plt.show()
```
