Bibliotecas necesarias para la ejecución

```python
import pyodbc 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

Definición de parámetros para la conexión con la base de datos

```python
conn_str_esea = (
    r'Driver={SQL Server};'
    r'Server=localhost\MSSQLSERVER01;'
    r'Database=csgo_esea;'
    r'Trusted_Connection=yes;'
    )
cnxn = pyodbc.connect(conn_str_esea)
```

Definición de query de obtención de datos

```python
query_cache = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_cache' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_dust = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_dust2' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_overpass = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_overpass' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_nuke = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_nuke' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_train = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_train' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_mirage = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_mirage' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_inferno = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_inferno' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
query_cobblestone = "SELECT [map],[winner_side],count([winner_side]) 'Número de victorias' FROM [csgo_esea].[dbo].[meta] where map='de_cbble' and (winner_side='CounterTerrorist' or winner_side='Terrorist') group by [map],[winner_side] order by map"
```

Carga de datos en dataframes

```python
data_cache = pd.read_sql_query(query_cache, cnxn)
data_dust = pd.read_sql_query(query_dust, cnxn)
data_overpass = pd.read_sql_query(query_overpass, cnxn)
data_nuke = pd.read_sql_query(query_nuke, cnxn)
data_train = pd.read_sql_query(query_train, cnxn)
data_mirage = pd.read_sql_query(query_mirage, cnxn)
data_inferno = pd.read_sql_query(query_inferno, cnxn)
data_cobblestone = pd.read_sql_query(query_cobblestone, cnxn)
```
